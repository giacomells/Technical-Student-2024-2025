```python
# Create a combined dataframe with an additional 'sextant' column
# First create a copy of each dataframe with the sextant number
df1_copy = df1.copy()
df1_copy['sextant'] = 1
df2_copy = df2.copy()
df2_copy['sextant'] = 2
df3_copy = df3.copy()
df3_copy['sextant'] = 3
df4_copy = df4.copy()
df4_copy['sextant'] = 4
df5_copy = df5.copy()
df5_copy['sextant'] = 5
df6_copy = df6.copy()
df6_copy['sextant'] = 6

# Concatenate all dataframes into a single one
df_combined = pd.concat([df1_copy, df2_copy, df3_copy, df4_copy, df5_copy, df6_copy], ignore_index=True)

# Create the same plot with the combined data
plt.figure(figsize=(12, 6))

# Use the same idx as before
idx = 454

# Create a list of position data and BPM names for each sextant
positions = []
bpm_names_all = []
bpm_masks = []

# Get the data for each sextant
for sextant in range(1, 7):
    # Get variable name for the current sextant
    var_name = f'BPMALPS_{sextant}:Orbit:positions'
    
    # Filter data for the current sextant
    df_sextant = df_combined[(df_combined.sextant == sextant) & 
                           (df_combined.nxcals_variable_name == var_name)]
    
    # Get the position data
    pos = df_sextant.iloc[idx]['nxcals_value']['elements']
    dim = df_sextant.iloc[idx]['nxcals_value']['dimensions']
    pos = np.array(pos.reshape(dim), dtype=float)
    positions.append(pos)
    
    # Get BPM names and masks for horizontal BPMs
    bpm_var = f'BPMALPS_{sextant}:Orbit:channelNames'
    bpm_names = df_combined[(df_combined.sextant == sextant) & 
                          (df_combined.nxcals_variable_name == bpm_var)].iloc[0]['nxcals_value']['elements']
    bpm_h_mask = ['.H' in name for name in bpm_names]
    bpm_names_all.append(bpm_names)
    bpm_masks.append(bpm_h_mask)
    
    # Plot the horizontal BPM positions at index 500
    plt.plot(bpm_names[bpm_h_mask], pos[1, 0][bpm_h_mask], label=f'Sextant {sextant}')

plt.xlabel('BPM Names')
plt.ylabel('Position at index 500')
plt.title('BPM Positions at Index 500 for Different Sextants')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
